```python
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

```python
#신경망과 같은 SVM 알고리즘은 데이터의 스케일에 매우 민감함. 보통은 특성에 따라 스케일을 조정해서 데이터를 변경
mglearn.plots.plot_scaling()
```

```python
'''
1번째 그래프 : 원본 데이터
2번째 그래프(StandardScaler) : 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 함
3번째 그래프(RobustScaler) : 평균과 분산 대신 중간값과 사분위 값(quartile)을 사용. 측정 에러등에 영향을 잘 안받음
4번째 그래프(MinMaxScaler) : 모든 특성이 정확하게 0과 1 사이에 위치하도록 데이터를 변경
5번째 그래프(Normalizer) : 특성 벡터의 유크리디안 길이가 1이 되도록 데이터 포인트 조정(반지름이 1인 원에 모든 데이터를 올림)
                            데이터의 크기 보다는 데이터의 방향이 중요할 때 사용하는 방법
'''
```

```python
#유방암 데이터에 scikit-learn을 적용해보자.
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()

X_train, X_test, y_train, y_test=train_test_split(cancer.data, cancer.target, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
#이 데이터에는 569개의 데이터 포인트가 있고, 각 데이터 포인트는 30개의 측정값으로 이루어짐. 이중 426개는 훈련, 143개는 테스트 세트
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
#데이터 변환. transform 메서드는 새로운 데이터 표현을 만들 때 사용
X_train_scaled=scaler.transform(X_train)
#스케일이 조정된 후의 데이터셋 속성 출력
print('변환된 후의 크기 : {}'.format(X_train.min(axis=0)))
print('스케일 조정 전 특성별 최솟값: {}'.format(X_train.min(axis=0)))
print('스케일 조정 전 특성별 최댓값: {}'.format(X_train.max(axis=0)))
print('스케일 조정 후 특성별 최솟값: {}'.format(X_train_scaled.min(axis=0)))
print('스케일 조정 후 특성별 최댓값: {}'.format(X_train_scaled.max(axis=0)))
```

```python
#변환된 데이터의 배열의 크기가 달라지지 않고, 데이터 크기가 변경되었을 뿐임
```

```python
#이 데이터에 SVM을 적용하려면 테스트 세트 또한 변환해야 함. transform 메서드에 X_test를 넣으면 됨
X_test_scaled=scaler.transform(X_test)
#스케일이 조정된 후 테스트 데이터의 속성
print('스케일 조정 후 특성별 최솟값:\n{}'.format(X_test_scaled.min(axis=0)))
print('스케일 조정 후 특성별 최댓값:\n{}'.format(X_test_scaled.max(axis=0)))
```

```python
#테스트 세트의 최소 최대가 0과 1이 아니라, 심지어 벗어난 경우도 있음.
#MinMaxScaler를 포함한 모든 스케일 모델은 훈련 세트와 테스트 세트에 같은 변환을 적용해야함.
from sklearn.datasets import make_blobs
#인위적인 데이터셋 생성
X, _=make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
#훈련 세트와 테스트 세트로 나눈다.
X_train, X_test=train_test_split(X,random_state=5, test_size=.1)

#훈련 세트와 테스트 세트의 산점도를 그린다.
fig, axes=plt.subplots(1,3,figsize=(13,4))
axes[0].scatter(X_train[:,0], X_train[:, 1], c=mglearn.cm2(0), label='train set', s=60)
axes[0].scatter(X_test[:,0], X_test[:,1], c=mglearn.cm2(1), label='test set', marker='^', s=60)
axes[0].legend(loc='upper left')
axes[0].set_title('original data')

#MinMaxScaler를 사용해 스케일 조정
scaler=MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

#스케일이 조정된 데이터의 산점도를 그린다
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2(0), label='train set', s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^', c=mglearn.cm2(1), label='test set', s=60)
axes[1].set_title('scaled data')

#테스트 스케일만 따로 조정한다.
#테스트 세트의 최솟값은 0, 최댓값은 1
#예시를 위한 것으로, 절대 아래와 같은 세팅을 하지 말 것
test_scaler=MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly=test_scaler.transform(X_test)

#잘못 조정된 데이터의 산점도
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2(0), label='train set', s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1], marker='^', c=mglearn.cm2(1), label='test set', s=60)
axes[2].set_title('Badly scaled data')

#plot in graph
for ax in axes:
    ax.set_xlabel('property 0')
    ax.set_ylabel('property 1')
```

```python
#보통 데이터의 fit 이후 transform을 바로 적용하는 경우가 많기 때문에 단축 메서드를 제공함
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
#보통은 이렇게
X_scaled=scaler.fit(X_train).transform(X_train)
#아래는 단축 메서드
X_scaled_d=scaler.fit_transform(X_train)
```

```python
#데이터의 전처리 효과가 가지는 의미
#먼저 원본 데이터로 SVC 모델을 만든다
from sklearn.svm import SVC

X_train, X_test, y_train, y_test=train_test_split(cancer.data, cancer.target, random_state=0)

svm=SVC(C=100)
svm.fit(X_train, y_train)
print('테스트 세트 정확도: {:.2f}'.format(svm.score(X_test, y_test)))
```

```python
#이번엔 SVC 모델을 학습시키기 전에 MinMaxScaler를 사용해 데이터의 스케일을 조정한다
#0~1사이로 스케일 조정
scaler=MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

#조정된 데이터로 SVM 학습
svm.fit(X_train_scaled, y_train)

#스케일 조정된 테스트 세트의 정확도
print('스케일 조정된 테스트 세트의 정확도: {:.2f}'.format(svm.score(X_test_scaled, y_test)))
```

```python
#조정된 데이터의 정확도가 더 높음을 알 수 있음
```

```python
#평균을 0, 분산을 1로 만드는 스케일 조정
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

#조정된 데이터로 SVM 학습
svm.fit(X_train_scaled, y_train)

#스케일 조정된 테스트 세트의 정확도
print('SVM 테스트 정확도: {:.2f}'.format(svm.score(X_test_scaled, y_test)))
```

```python
#비지도 학습을 통해 데이터를 변환하는 이유는 여러가지가 있지만, 보통은 시각화하거나, 데이터를 압축하거나, 추가적인 처리를 위해 사용
#이런 용도로 가장 간단하고 흔히 사용하는 알고리즘에는 주성분 분석(PCA, Principal Component Analysis)이 있음
#이외에도 비음수 행렬 분해(NMF, Non-negative Matrix Factorization)도 있음
#또 2차원 산점도를 이용해 시각화 용도로 사용하는 t-SNE(t-distributed Stochastic Neighbor Embeding) 알고리즘도 있음
```

```python
#PCA
#PCA는 특성들이 통계적으로 상관관계가 없도록 데이터셋을 회전시키는 기술.
mglearn.plots.plot_pca_illustration()
```

```python
'''
첫 번째 그래프: 원본 데이터. 성분 1이라고 쓰여있는 분산이 가장큰 방향을 가짐. 즉, 이 벡터가 데이터에서 가장 많은 정보를 가짐.
즉, 데이터간 상관관계가 가장 큼. 데이터의 주된 성분의 방향을 주성분이라고 함. 일반적으로 원본 특성 개수만큼의 주성분이 존재함.

두 번째 그래프: 같은 데이터지만 주성분 1과 2를 x축솨 y축에 나랗나도록 회전시킨 것. 회전하기 전에 데이터에서 평균을 빼서 중심은 원점.
correlation matrix를 곱하여 회전시킨다.

세 번째 그래프: 첫 번째 주 성분만 유지하기 위해 차원 축소 용도로 사용도 가능. 가장 유용한 방향을 찾아서 그 방향의 성분을 유지하는 것

네 번째 그래프: 데이터에서 평균을 더해서 반대로 회전. 데이터 포인트들은 원래 특성 공간에 놓여있지만, 첫 번째 주 성분의 정보를 담음.
이 방식은 데이터에서 노이즈를 제거하거나 주성분에서 유지되는 정보를 시각화하는데 사용됨.
'''
```

```python
#PCA를 적용해 유방암의 데이터셋을 시각화한다.
#유방암 데이터셋은 30개의 특성을 가지고 있기 때문에, 이 산점도를 다 그리는 미친짓은 하기 힘듬.
#양성과 악성 두 클래스에 대해 각 특성의 히스토그램을 그려보자.

fig, axes=plt.subplots(15,2,figsize=(10,20))
malignant=cancer.data[cancer.target==0]
benign=cancer.data[cancer.target==1]

ax=axes.ravel()

for i in range(30):
    _, bins=np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel('component size')
ax[0].set_ylabel('frequency')
ax[0].legend(['malignant', 'benign'], loc='best')
fig.tight_layout()
```

```python
#smoothness error 특성은 두 히스토그램이 겹쳐 별 쓸모가 없으나, worst concave points는 확실히 구분되어 유용한 특성
#하지만 이 히스토그램만 가지고는 특성들 간의 상관관계를 알 수 없음. PCA를 사용하면 주요 상호작용을 찾아낼 수 있다.
#PCA를 적용하기 전에 StandardScaler를 사용해 각 특성의 분산을 1로 만든다.
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()

scaler=StandardScaler()
scaler.fit(cancer.data)
X_scaled=scaler.transform(cancer.data)
```

```python
#PCA를 적용한다.
from sklearn.decomposition import PCA
#데이터의 첫 두 개의 주성분만 유지한다.
pca=PCA(n_components=2)
#유방암 데이터로 PCA 모델을 만든다.
pca.fit(X_scaled)

#처음 두 개의 주성분을 사용해 데이터를 변환한다.
X_pca=pca.transform(X_scaled)
print('원본 데이터 형태: {}'.format(str(X_scaled.shape)))
print('축소된 데이터 형태: {}'.format(str(X_pca.shape)))
```

```python
#그 후 맨 처음 두 개의 주성분을 그린다.
#클래스를 색깔별로 구분하여 처음 두 개의 주성분을 그래프로 나타낸다.
plt.figure(figsize=(8,8))
mglearn.discrete_scatter(X_pca[:,0], X_pca[:,1], cancer.target)
plt.legend(['malignant', 'benign'], loc='best')
plt.gca().set_aspect('equal')
plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
```

```python
#PCA로는 그 두 개의 주성분이 뭔지 파악하기가 쉽지 않다는 단점이 있음. PCA 객체에 components_성분에 주성분이 저장됨
print('PCA 주성분 형태: {}'.format(pca.components_.shape))
```

```python
#components_의 각 행은 주성분 하나씩을 나타내며, 중요도에 따라 정려로디어있음. 열은 원본 데이터의 특성에 대응하는 값.
#components_의 값을 출력해보자.
print('PCA 주성분: {}'.format(pca.components_))
```

```python
#히트맵으로 시각화하여 더 쉽게 이해해보자.
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0,1], ['1st principal component', '2nd principal component'])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)), cancer.feature_names, rotation=60, ha='left')
plt.xlabel('property')
plt.ylabel('principal component')
```

```python
#첫 번째 주 성분은 모든 부호가 같으므로 공통의 상호관계가 있음을 확인할 수 있지만, 두 번째 주 성분은 그렇지 않아 의미를 찾기 힘듬.
```

```python
#고유 얼굴(eigenface) 특성 추출
#PCA는 특성 추출에도 이용됨. PCA를 이용해 LFW(Labeled Faces in the World) 데이터셋의 얼굴 이미지에서 특성을 추출해보자.
#흑백 이미지를 사용하고, 스케일을 줄였다.
#아래의 이미지는 몇개의 사진 샘플
from sklearn.datasets import fetch_lfw_people
people=fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape=people.images[0].shape

fig, axes=plt.subplots(2,5,figsize=(15,8),subplot_kw={'xticks': (), 'yticks': ()})
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
#LFW 데이터셋에는 62명의 얼굴을 찍은 이미지가 총 3023개 있으며, 각 이미지의 크기는 87*65픽셀
print('people.images.shape: {}'.format(people.images.shape))
print('클래스 개수: {}'.format(len(people.target_names)))
```

```python
#이 데이터셋은 조지 부시와 콜린 파워에 조금 편중되어있다. 이를 확인해보자.
#각 타깃이 나타난 횟수 계산
counts=np.bincount(people.target)
#타깃별 이름과 횟수 출력
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print('{0:25} {1:3}'.format(name, count), end='    ')
    if (i+1)%3==0:
        print()
```

```python
#데이터셋의 편중을 막기 위해 사람마다 50개의 이미지만 선택한다.
mask=np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]]=1

X_people=people.data[mask]
y_people=people.target[mask]

#0~255 사이의 흑백 이미지의 픽셀 값을 0~1 스케일로 조정한다.
#MinMaxScaler와 거의 같은 효과를 가짐
X_people=X_people/255.
```

```python
#분류하려는 얼굴과 가장 비슷한 얼굴 이미지를 찾는 1-최근접 이웃 분류기를 사용해본다.
from sklearn.neighbors import KNeighborsClassifier
#훈련 세트와 테스트 세트로 나눈다.
X_train, X_test, y_train, y_test=train_test_split(X_people, y_people, stratify=y_people, random_state=0)
#이웃 개수를 1개로 하여 KneighborsClassifier 모델을 만든다.
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print('1-최근접 이웃의 테스트 세트 점수 : {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
#별로 좋은 모델은 아니다. PCA를 활용해보자. PCA 화이트닝(whitening) 옵션을 사용해 주성분의 스케일이 같아지도록 조정한다.
mglearn.plots.plot_pca_whitening()
```

```python
#PCA 객체를 훈련 데이터로 학습시켜 첫 100개의 주성분을 추출한다. 그 후 훈련 데이터와 테스트 데이터를 변환한다.
pca=PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)

print('X_train_pca.shape: {}'.format(X_train_pca.shape))
```

```python
#이제 여기에 1-최근접 이웃을 이미지로 분류해보자.
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print('테스트 세트 정확도: {:.2f}'.format(knn.score(X_test_pca, y_test)))
```

```python
#좀 더 나아진 모습을 볼 수 있음
#몇가지 주성분을 시각화해본다.
print('pca.components_.shape: {}'.format(pca.components_.shape))
```

```python
#비음수 행렬 분해, NMF(Non-negative matrix factorization)
#PCA에서는 데이터의 분산이 갖아 크고 수직인 성분을 찾았다면, NMF에서는 음수가 아닌 성분과 계수 값을 찾음.
#주성분과 계수가 모두 0보다 크거나 같아야 함. 따라서 음수가 아닌 특성을 가진 데이터에만 적용할 수 있음
#인위적인 데이터에 NMF를 적용해보자.

mglearn.plots.plot_nmf_illustration()
```

```python
#얼굴 이미지에 NMF를 적용해보자.
mglearn.plots.plot_nmf_faces(X_train, X_test, image_shape)
```

```python
#PCA에 비해선 더 흐릿한 것을 볼 수 있는데, PCA가 재구성 측면에서 최선의 방향을 찾기 때문.
#NMF는 데이터를 인코딩하거나 재구성하기보다는 데이터의 유용한 패턴을 찾는데 활용된다.
#성분을 15개만 추출해본다.
from sklearn.decomposition import NMF
nmf=NMF(n_components=15, random_state=0)
nmf.fit(X_train)
X_train_nmf=nmf.transform(X_train)
X_test_nmf=nmf.transform(X_test)

fig, axes=plt.subplots(3,5,figsize=(15,12), subplot_kw={'xticks': (), 'yticks': ()})

for i, (component, ax) in enumerate(zip(nmf.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape))
    ax.set_title('component {}'.format(i))
```

```python
#PCA보단 조금 더 얼굴 원형처럼 보임(성분이 모두 양수이기 때문) 성분 3, 7이 특히 강하게 나타난 사진을 찾아보자.
compn=3
#4번째 성분으로 정렬하여 처음 10개 이미지를  출력한다.
inds=np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes=plt.subplots(2,5,figsize=(15,8), subplot_kw={'xticks': (), 'yticks': ()})

for i, (ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))

compn=7
#8번째 성분으로 정렬하여 처음 10개 이미지를 출력한다.
inds=np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes=plt.subplots(2,5,figsize=(15,8), subplot_kw={'xticks': (), 'yticks': ()})

for i, (ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))
```

```python
#성분 3의 계수 값이 큰 얼굴들은 왼쪽으로 돌아가 있는 것을 볼 수 있음.
#NMF는 소리, 유전자 표현, 텍스트 데이터처럼 덧붙이는 구조를 가진 데이터네 적합함.
#인위적인 데이터셋을 이용해 확인해보자.
S=mglearn.datasets.make_signals()
plt.figure(figsize=(6,1))
plt.plot(S,'-')
plt.xlabel('time')
plt.ylabel('signal')
```

```python
#합쳐진 신호에서 원본 신호를 복원해야 한다. 여러 대의 관측 장치가 일련의 측정 데이터를 제공한다고 하자.
#원본 데이터를 사용해 100개의 측정 데이터를 만든다.
A=np.random.RandomState(0).uniform(size=(100,3))
X=np.dot(S,A.T)
print('측정 데이터의 형태: {}'.format(X.shape))
```

```python
#NMF를 사용해 3개의 신호를 복원한다.
nmf=NMF(n_components=3, random_state=42)
S_=nmf.fit_transform(X)
print('복원한 신호 데이터 형태: {}'.format(S_.shape))
```

```python
#비교를 위해 PCA도 적용해보자.
pca=PCA(n_components=3)
H=pca.fit_transform(X)
```

```python
#NMF와 PCA로 찾은 신호를 그림으로 나타낸다.
models=[X,S,S_,H]
names=['measured signal', 'original signal', 'NMF signal', 'PCA signal']

fig, axes=plt.subplots(4,figsize=(8,4), gridspec_kw={'hspace': .5}, subplot_kw={'xticks': (), 'yticks': ()})

for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:,:3], '-')
```

```python
#NMF가 가장 유사한 데이터 신호를 낸 것을 볼 수 있음. 이외에도 여러 패턴 추출이 있으므로, ICA, FA, 희소 코딩 등의 분해 메서드를 검색하자.
```

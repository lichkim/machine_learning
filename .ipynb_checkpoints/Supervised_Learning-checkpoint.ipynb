```python
'''
지도 학습에는 분류(classfication)와 회귀(regression)이 있음


분류 : 미리 정의된, 가능성 있는 여러 클래스 레이블 중 하나를 예측하는 것(이산적임)
    
    -이진 분류(binary classfication) : 딱 두개의 클래스로 구분하는 이진분류
    -다중 분류(multiclass classification) : 셋 이상의 클래스로 구분하는 분류
    
회귀 : 실수를 예측. 예를 들어, 어떤 사람의 교육 수준, 나이, 주거지 등을 바탕으로 연간 소득을 예측하는 것(연속적임)


모델이 처음 보는 데이터에 대해 정확하게 예측할 수 있으면 이를 훈련 세트에서 테스트 세트로 일반화(generalization)됐다고 함.

    -과대적합(overfitting) : 모델이 훈련세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때
    -과소적합(underfitting) : 모델이 훈련세트의 각 샘플에서 너무 멀리 맞춰져서 너무 간단한 모델이 될 때

-->다양한 데이터 포인트가 많을 수록 과대적합 없이 더 복잡한 모델을 만들 수 있음. 비슷한 데이터를 모으는것은 많이 도움되지 않음
'''
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

```python
#예제에 사용할 데이터셋

#데이터셋을 만든다.
X,y=mglearn.datasets.make_forge()

#산점도를 그린다.
mglearn.discrete_scatter(X[:, 0],X[:, 1],y)
plt.legend(["class 0, class 1"], loc=4)
plt.xlabel('property 1')
plt.ylabel('property 2')
#X의 shape은 (26,2)이므로, 데이터 포인트 26개와 특성 2개를 가짐
print("X.shape: {}".format(X.shape))
```

```python
#회귀 알고리즘에는 인위적으로 만든 wave 데이터셋 사용.wave 데이터셋은 입력 특성 하나와 모델링할 타깃 변수를 가짐.
X,y=mglearn.datasets.make_wave(n_samples=40)
plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel('property')
plt.ylabel('target')
```

```python
'''
특성이 적은 데이터셋이(저차원 데이터셋, property가 낮은 왼쪽의 데이터들)이
특성이 많은 데이터셋(고차원 데이터셋, property가 높은 오른쪽의 데이터들)과 비슷한 분포를 그리지 않음을 알 수 있음.
그러므로, 실제 알고리즘을 사용할 때에는 특성이 적은 저차원 데이터셋을 사용하는 것이 바람직함
'''
```

```python
'''
이번 지도 학습에 사용할 예제는 scikit-learn에 들어있는 실제 데이터셋인 위스콘신 유방암 데이터셋과 조직 데이터.
각 종양은 양성 종양 혹은 악성 종양으로 레이블되어있고, 조직 데이터를 기반으로 종양이 악성인지를 예측하는 것이 목표
'''

from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
print('cancer.keys(): \n{}'.format(cancer.keys()))
```

```python
#569개의 데이터 포인트와 30개의 특성을 가짐을 알 수 있음
print('유방암 데이터의 형태: {}'.format(cancer.data.shape))
```

```python
#이중 212개는 악성, 357개는 양성임
print('클래스별 샘플 개수:\n{}'.format({n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
#python의 zip 함수는 개수가 같은 iterable한 자료형을 묶어주는 역할. 예를 들어 list(zip([1,2,3],[4,5,6]))=[(1,4),(2,5),(3,6)]
```

```python
#feature_names 속성을 확인하여 각 특성의 의미를 알 수 있음
print('특성 이름:\n{}'.format(cancer.feature_names))
#데이터에 대한 더 자세한 정보는 cancer.DESCR에서 확인 가능
```

```python
#회귀 분석용 데이터셋으로는 보스턴 주택가격 데이터셋을 사용. 범죄율, 강과의 인접도, 고속도로 접근성 등의 정보로 주택 가격 예측
#이 데이터엔 506개의 데이터 포인트와 13개의 특성이 있음
from sklearn.datasets import load_boston
boston=load_boston()
print('데이터의 형태 : {}'.format(boston.data.shape))
```

```python
#특성끼리 곱하여(혹은, 상호작용) 범죄율*고속도로 접근성 까지도 특성으로 생각하는 특성 공학을 사용한 데이터셋을 사용한다.
X,y=mglearn.datasets.load_extended_boston()
print('X.shape: {}'.format(X.shape))
#13H2=104개의 특성
```

```python
#k-최근접 이웃(k-NN Algorithm)
#가장 가까운 훈련 데이터 포인트 하나를 최근접 이웃으로 찾는 것을 k-NN 알고리즘이 수행한다.
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
'''
위의 그림은 추가한 3개의 데이터 포인트(별모양)에서 가장 가까운 훈련 데이터를 연결한 그림. 1-NN 알고리즘이 된다.
1개가 아니라, k개의 최근접 이웃을 고를 수도 있음. 아래의 그림은 3-NN 알고리즘이 될 것이다.
'''
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
#훈련 세트와 테스트 세트를 나눈다.
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#clf : KNeighborsClassifier 클래스 객체
clf=KNeighborsClassifier(n_neighbors=3)
```

```python
#학습. KNeighborsClassifier 클래스에서 학습은 예측 시 이웃을 찾을 수 있도록 데이터를 저장하는 것
clf.fit(X_train, y_train)
```

```python
#테스트 데이터에 대한 예측. 테스트 세트의 각 데이터 포인트에 대해 훈련 세트에서 가장 가까운 이웃을 계산한다.
print('테스트 세트 예측: {}'.format(clf.predict(X_test)))
```

```python
#모델 예측 평가
print('테스트 세트 정확도: {:.2f}'.format(clf.score(X_test, y_test)))
```

```python
#KNeighborsClassifier 분석
#데이터가 모두 2차원 데이터셋이므로, 가능한 모든 테스트 포인트의 예측을 xy평면에 그릴 수 있음. 이를 통해 결정 경계를 볼 수 있음
#아래의 예제는 1-NN, 3-NN, 9-NN 의 결정 경계를 보여줌

fig, axes=plt.subplots(1,3,figsize=(10,3))

for n_neighbors, ax in zip([1,3,9], axes):
    #fit 메서드는 self 객체를 반환하여 객체 생성과 fit 메서드를 한 줄에 쓰는게 가능
    clf=KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=0.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} Neighbor".format(n_neighbors))
    ax.set_xlabel('property 0')
    ax.set_ylabel('property 1')
axes[0].legend(loc=3)
```

```python
#이웃의 수를 몇개로 잡느냐에 따라 underfitting이 될 수도 있고, overfitting이 될 수도 있음
```

```python
#일전의 유방암 데이터셋에 적용시켜본다.
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy=[]
test_accuracy=[]
#1~10까지 n_neighbors를 적용
neighbors_settings=list(range(1,11))

for n_neighbors in neighbors_settings:
    #모델 생성
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    #훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label='training accuracy')
plt.plot(neighbors_settings, test_accuracy, label='test accuracy')
plt.ylabel('accuracy')
plt.xlabel('n_neighbors')
plt.legend()
plt.show()
```

```python
#이웃을 6개로 설정했을 때가 가장 테스트 케이스가 정확도가 높은것을 보아 정확한 것을 알 수 있음
```

```python
#wave 데이터셋을 이용하여 이웃이 하나인 최근접 이웃을 사용한다.
mglearn.plots.plot_knn_regression(n_neighbors=1)
```

```python
#이웃이 1개일 때, 테스트 예측은 훈련된 데이터아 가장 가까운 데이터에 향함
```

```python
#이웃이 2개 이상일 땐, 이웃간의 평균이 예측이 됨
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

```python
#회귀분석용 최근접 이웃 알고리즘은 KNeighborsRegressor 클래스에 구현되어 있음
from sklearn.neighbors import KNeighborsRegressor

X, y=mglearn.datasets.make_wave(n_samples=40)

#wave 데이터셋을 훈련 세트와 테스트 세트로 나눔
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=0)

#이웃의 수를 3으로 설정하여 모델의 객체를 만듬
reg=KNeighborsRegressor(n_neighbors=3)
#훈련 데이터와 타깃으로 모델 학습
reg.fit(X_train, y_train)
```

```python
#테스트 세트 예측
print('테스트 세트 예측:\n{}'.format(reg.predict(X_test)))
```

```python
#score 메서드를 사용해 모델 평가
#R^2는 회귀 모델에서 예측의 정확도를 0~1 사이의 값으로 계산하는 것. 1-(sum((y-y_hat)^2)/sum((y-avg(y))^2), y_hat은 예측값
print('테스트 세트 R^2: {:.2f}'.format(reg.score(X_test, y_test)))
```

```python
#KNeighborsRegressor 분석
fig, axes=plt.subplots(1,3,figsize=(15,4))
#-3과 3 가이에 1000개의 데이터 포인트를 만든다
line=np.linspace(-3,3,1000).reshape(-1,1)
for n_neighbors, ax in zip([1,3,9], axes):
    #1, 3, 9 이웃을 사용한 예측
    reg=KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
    '{} neighbors training score: {:.2f} test score: {:.2f}'.format(n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel('property')
    ax.set_ylabel('target')
    axes[0].legend(['model prediction', 'training data/target', 'test data/target'], loc='best')
```

```python
#이웃을 많이 사용할 수록 훈련 데이터와는 잘 안맞을 수도 있으나 더 안정된 곡선을 얻을 수 있음
```

```python
'''
회귀의 선형 모델
predict(y)=w[0]x[0]+w[1]x[1]+w[2]x[2]+...+b (b:bias w:weight, x:data 신경망이랑 똑같다.)
1차원 wave 데이터셋을 회귀 선형 모델을 통해 분석하자.
'''
mglearn.plots.plot_linear_regression_wave()
```

```python
#특성이 1개일 땐 직선, 2개일 땐 평면... 같은 식으로 특성의 갯수=차원이 됨. 훈련 데이터 보다 특성이 더 많은 경우엔 선형함수로 모델 가능.
#회귀를 위한 선형 모델은 w와 b를 학습하는 방법과 모델의 복잡도를 제어하는 방법에서 다양하게 있음.
#다음에는 가장 인기있는 선형 모델
```

```python
#선형회귀(최소제곱법)
'''
최소제곱법이란, 아래의 공식에서 구한 에러에서 이 에러를 최소화하는 파라미터 w와 b를 찾는 과정을 말한다.
매개변수가 없다는 장점이 있지만, 모델의 복잡도를 제어할 방법 또한 없다는 것이 단점.
      
MSE = sum(for all y){(y-target)^2}
'''

from sklearn.linear_model import LinearRegression
X,y=mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=42)

lr=LinearRegression().fit(X_train, y_train)
```

```python
#weight(coefficient)는 lr 객체의 coef_ 속성에 저장되어 있고, bias는 intercept_ 속성에 저장되어 있음
print('lr.coef_: {}'.format(lr.coef_))
print('lr.intercept: {}'.format(lr.intercept_))
```
